```python
#handling of the missing values
data['Lat'].fillna(data['Lat'].mean(),inplace=True)
data['Long'].fillna(data['Long'].mean(),inplace=True)
data['WHO Region'].fillna(data['WHO Region'].mode()[0],inplace=True)

# Remove rows with any missing values
# data.dropna(inplace=True)

# Remove columns with any missing values
# data.dropna(axis=1, inplace=True)

# Fill missing values with mean
# data.fillna(data.mean(), inplace=True)
```

```python
#changing the data type
data['Long']=data['Long'].astype('string')
data['Date']=pd.to_datetime(data['Date'])
```
