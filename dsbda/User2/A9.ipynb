```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Iris.csv')
data.head(),data.shape
```

```python
# (1) List down the features and their types:
# The Iris dataset consists of four numeric features and one nominal target variable.

# Sepal Length (numeric)
# Sepal Width (numeric)
# Petal Length (numeric)
# Petal Width (numeric)
# Species (nominal)
```

```python
# Plot histograms for each feature
data.drop('Id', axis=1).hist(figsize=(10, 8))
plt.show()
```

```python
# Plot boxplots for each feature
numeric_value=[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]

plt.figure(figsize=(10, 8))
sns.boxplot(data=data.drop(['Id', 'Species'], axis=1))
plt.title('Boxplot for each Feature')
plt.show()
```
