```python
# Data Visualization III
# Download the Iris flower dataset or any other dataset into a DataFrame. (e.g.,
# https://archive.ics.uci.edu/ml/datasets/Iris ). Scan the dataset and give the inference as:
# 1. List down the features and their types (e.g., numeric, nominal) available in the dataset.
# 2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
# 3. Create a boxplot for each feature in the dataset.
# 4. Compare distributions and identify outliers
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
df = df.drop(columns=["Id"])
df.head()
```

```python
df.describe()
```

```python
# Unique species
df.Species.unique()
```

```python
# Count of Species
sns.countplot(x='Species', data=df)
plt.show()
```

```python
# Data Types
df.dtypes
```

```python
# Pairplot
sns.pairplot(data=df,hue="Species")
plt.show()
```

```python
# Histogram to illustarte feature distrivbutions
```

```python
sns.histplot(x=df["SepalLengthCm"],bins=7)
plt.title("Sepal Length")
sns.displot(data=df,hue="Species",x="SepalLengthCm",kde=True,bins=7)
plt.show()
```

```python
sns.histplot(x=df["SepalWidthCm"],bins=10)
plt.title("Sepal Width")
sns.displot(data=df,hue="Species",x="SepalWidthCm",kde=True,bins=10)
plt.show()
```

```python
sns.histplot(x=df["PetalLengthCm"],bins=7)
plt.title("Petal Length")
sns.displot(data=df,hue="Species",x="PetalLengthCm",kde=True,bins=7)
plt.show()
```

```python
sns.histplot(x=df["PetalWidthCm"],bins=10)
plt.title("Petal Width")
sns.displot(data=df,hue="Species",x="PetalWidthCm",kde=True,bins=10)
plt.show()
```

```python
### Boxplots of categorical variables with other numerical variables
#
plt.figure(figsize=(15,15))
plt.subplot(221)
sns.boxplot(x="Species", y='SepalLengthCm', data=df)
plt.subplot(222)
sns.boxplot(x="Species", y='SepalWidthCm', data=df)
plt.subplot(223)
sns.boxplot(x="Species", y='PetalLengthCm', data=df)
plt.subplot(224)
sns.boxplot(x="Species", y='PetalWidthCm', data=df)
plt.show()
```

```python
# Outliers
plt.figure(figsize=(12,12))
plt.subplot(221)
sns.boxplot(x='SepalLengthCm', data=df)
plt.title("Sepal Length")
plt.subplot(222)
sns.boxplot(x='SepalWidthCm', data=df)
plt.title("Sepal Width")
plt.subplot(223)
sns.boxplot(x='PetalLengthCm', data=df)
plt.title("Petal Length")
plt.subplot(224)
sns.boxplot(x='PetalWidthCm', data=df)
plt.title("Petal Width")
plt.show()
```
